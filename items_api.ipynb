```python
# import relevant libraries
import pandas as pd
import requests
import time

url_sql = "https://api.opendota.com/api/explorer"
```

```python
test = "SELECT * FROM player_matches WHERE match_id BETWEEN 3710610589 AND 5000000000 LIMIT 5"

params = {"sql": test}

response = requests.get(url=url_sql, params=params)

print(response)

data = response.json()

data

df = pd.DataFrame(data["rows"])

# n = 
df
```

```python
n = 3710610588

df_players_matches_items = pd.DataFrame()

while n < 8721721938:
  test = f"SELECT match_id, account_id, hero_id, item_0, item_1, item_2, item_3, item_4, item_5 FROM player_matches WHERE match_id BETWEEN {n} AND {n + 75000000}"
  params = {"sql": test}
  n += 75000000
  response = requests.get(url=url_sql, params=params)

  data = response.json()
  df_temp = pd.DataFrame(data["rows"])

  df_players_matches_items = pd.concat([df_players_matches_items, df_temp], axis=0)
  time.sleep(2)
  print(df_players_matches_items.shape)

df_players_matches_items.shape
```
