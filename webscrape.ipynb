```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd

url = "https://www.dota2.com/hero/"

heroes_df = pd.read_csv("data/raw/hero_names.csv")

df = pd.DataFrame()

hero_names = heroes_df["hero"].to_list()

hero_names
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

for hero_name in hero_names:

  driver.get(url + hero_name)

  time.sleep(4)

  soup = BeautifulSoup(driver.page_source, "html.parser")
  character_dict = {}
  character_soup = soup.find(name='div', attrs={"class": "_2jP25gUPi28sj4muJVFYir"})

  char_attr_img = character_soup.find(name="img", attrs={"class": "_1A3VDYOIxIzxr-UZr9KAbs"})["src"]
  char_attr_txt = character_soup.find(name="div", attrs={"class": "_3HGWJjSyOjmlUGJTIlMHc_"}).getText()
  char_name = character_soup.find(name="div", attrs={"class": "_2IcIujaWiO5h68dVvpO_tQ"}).getText()
  char_blurb = character_soup.find(name="div", attrs={"class": "_2r7tdOONJnLw_6bQuNZj5b"}).getText()
  char_history = character_soup.find(name="div", attrs={"class": "_33H8icML8p8oZrGPMaWZ8o"}).getText()
  char_type_img = character_soup.find(name="img", attrs={"class": "_9eyi-y0N7HQeAgY_e4IX3"})["src"]
  char_type = character_soup.find(name="div", attrs={"class": "_3ce-DKDrVB7q5LsGbJdZ3X"}).getText()

  char_abilities = character_soup.find_all(name="div", attrs={"class": "_3ty-o0Vg7AspURxYD_vDO0"})

  for i in range(1, len(char_abilities)):
    # print(char_abilities[i].getText())
    abil_num = "abil" + str(i)
    #title
    abil_title = char_abilities[i].find(name="div", attrs={"class": "oRzqVIPO5vD9aPxv582G9"}).getText()
    abil_text = char_abilities[i].find(name="div", attrs={"class": "_1GUxaKU8ts0aIJaMfIfMS_"}).getText()
    abil_img = char_abilities[i].find(name="img", attrs={"class": "_3Chop4A9yz7Af_BwR1r_NW"})["src"]
    abil_vid = char_abilities[i].find(name="video", attrs={"class": "_22nJ5nsfHDS2jEscPEne0-"}).find_all("source")[0]["src"]

    character_dict.update({abil_num + "title": abil_title
                          , abil_num + "text": abil_text
                          , abil_num + "img": abil_img
                          , abil_num + "vid": abil_vid
                          })

  character_dict.update({"char_attr_img": char_attr_img,
                        "char_attr_txt": char_attr_txt,
                        "char_name": char_name,
                        "char_blurb": char_blurb,
                        "char_history": char_history,
                        "char_type_img": char_type_img,
                        "char_type": char_type,
                        })

  temp_df = pd.DataFrame(character_dict, index=[0])
  df = pd.concat([df, temp_df], axis=0)

  # time.sleep(1)
  df

driver.quit()
```
